```python
import requests
from bs4 import BeautifulSoup

def scrape_links(url):
    """
    Scrape website links from the given URL.
    """
    # Send a GET request to the URL
    page = requests.get(url)

    # Check if the request was successful (status code 200)
    if page.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(page.content, 'html.parser')

        # Find all <a> tags containing links
        links = soup.find_all('a', href=True)

        # Extract and return the links
        website_links = set()  # Use a set to store unique links
        for link in links:
            href = link['href']
            if href.startswith('https://fallout.fandom.com/wiki/') and not href.endswith('.txt'):
                website_links.add(href)
        return list(website_links)  # Convert the set back to a list before returning
    else:
        # Print an error message if the request was not successful
        print("Failed to retrieve the webpage - Status code:", page.status_code)
        return []

```

```python
# Create the folder if it doesn't exist
folder_name = "Nukapedia"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
len(unique_links)
```

```python
# Iterate over each website in the list
for website in unique_links:
    # Send a GET request to the URL
    page = requests.get(website)

    # Check if the request was successful (status code 200)
    if page.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(page.content, 'html.parser')

        # Find the specific element containing the desired content (you can adjust this based on the structure of the webpage)
        content_div = soup.select_one('#mw-content-text > div')

        # Extract the text content from the element
        result = content_div.get_text()

        # Save the content to a text file in the folder
        filename = os.path.join(folder_name, website.split("/")[-1] + ".txt")  # Extract the filename from the URL
        with open(filename, 'w') as file:
            file.write(result)

        print("Content from", website, "saved to", filename)
        print("-" * 50)  # Separator for readability
    else:
        # Print an error message if the request was not successful
        print("Failed to retrieve", website, "- Status code:", page.status_code)
```
